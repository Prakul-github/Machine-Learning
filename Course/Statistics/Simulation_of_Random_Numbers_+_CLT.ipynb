```python
#Central Limit Theorem - Normal Population
a=[]
b=[]
c=[]
d=[]
e=[]
f=[]
for i in range(10000):
  X_rand = np.random.normal(loc = 10, scale = 5, size = 1)
  a.append(sum(X_rand)/len(X_rand))
  Y_rand = np.random.normal(loc = 10, scale = 5, size = 5)
  b.append(sum(Y_rand)/len(Y_rand))
  Z_rand = np.random.normal(loc = 10, scale = 5, size = 10)
  c.append(sum(Z_rand)/len(Z_rand))
  U_rand = np.random.normal(loc = 10, scale = 5, size = 30)
  d.append(sum(U_rand)/len(U_rand))
  V_rand = np.random.normal(loc = 10, scale = 5, size = 50)
  e.append(sum(V_rand)/len(V_rand))
  W_rand = np.random.normal(loc = 10, scale = 5, size = 100)
  f.append(sum(W_rand)/len(W_rand))

plt.hist(a, bins = 100, label = 'simulated', alpha = 0.5, density = True)
plt.hist(b, bins = 100, label = 'simulated', alpha = 0.5, density = True)
plt.hist(c, bins = 100, label = 'simulated', alpha = 0.5, density = True)
plt.hist(d, bins = 100, label = 'simulated', alpha = 0.5, density = True)
plt.hist(e, bins = 100, label = 'simulated', alpha = 0.5, density = True)
plt.hist(f, bins = 100, label = 'simulated', alpha = 0.5, density = True)
plt.show
```

```python
#Central Limit Theorem - Uniform Population
a=[]
b=[]
c=[]
d=[]
e=[]
f=[]
for i in range(100000):
  X_rand = np.random.uniform(low = 10, high = 50, size = 1)
  a.append(sum(X_rand)/len(X_rand))
  Y_rand = np.random.uniform(low = 10, high = 50, size = 5)
  b.append(sum(Y_rand)/len(Y_rand))
  Z_rand = np.random.uniform(low = 10, high = 50, size = 10)
  c.append(sum(Z_rand)/len(Z_rand))
  U_rand = np.random.uniform(low = 10, high = 50, size = 30)
  d.append(sum(U_rand)/len(U_rand))
  V_rand = np.random.uniform(low = 10, high = 50, size = 50)
  e.append(sum(V_rand)/len(V_rand))
  W_rand = np.random.uniform(low = 10, high = 50, size = 100)
  f.append(sum(W_rand)/len(W_rand))

plt.hist(a, bins = 50, label = 'simulated', alpha = 0.5, density = True)
plt.hist(b, bins = 50, label = 'simulated', alpha = 0.5, density = True)
plt.hist(c, bins = 50, label = 'simulated', alpha = 0.5, density = True)
plt.hist(d, bins = 50, label = 'simulated', alpha = 0.5, density = True)
plt.hist(e, bins = 50, label = 'simulated', alpha = 0.5, density = True)
plt.hist(f, bins = 50, label = 'simulated', alpha = 0.5, density = True)
plt.show
```

```python
#Central Limit Theorem - Poisson Population
a=[]
b=[]
c=[]
d=[]
e=[]
f=[]
for i in range(10000):
  X_rand = np.random.poisson(50, size = 1)
  a.append(sum(X_rand)/len(X_rand))
  Y_rand = np.random.poisson(50, size = 5)
  b.append(sum(Y_rand)/len(Y_rand))
  Z_rand = np.random.poisson(50, size = 10)
  c.append(sum(Z_rand)/len(Z_rand))
  U_rand = np.random.poisson(50, size = 30)
  d.append(sum(U_rand)/len(U_rand))
  V_rand = np.random.poisson(50, size = 50)
  e.append(sum(V_rand)/len(V_rand))
  W_rand = np.random.poisson(50, size = 100)
  f.append(sum(W_rand)/len(W_rand))

plt.hist(a, bins = 40, label = 'simulated', alpha = 0.5, density = True)
plt.hist(b, bins = 40, label = 'simulated', alpha = 0.5, density = True)
plt.hist(c, bins = 40, label = 'simulated', alpha = 0.5, density = True)
plt.hist(d, bins = 40, label = 'simulated', alpha = 0.5, density = True)
plt.hist(e, bins = 40, label = 'simulated', alpha = 0.5, density = True)
plt.hist(f, bins = 40, label = 'simulated', alpha = 0.5, density = True)
plt.show
```

```python
#Central Limit Theorem - Bernoulli Population
a=[]
b=[]
c=[]
d=[]
e=[]
f=[]
for i in range(10000):
  X_rand = np.random.binomial(1, 0.8, size = 1)
  a.append(sum(X_rand)/len(X_rand))
  Y_rand = np.random.binomial(1, 0.8, size = 5)
  b.append(sum(Y_rand)/len(Y_rand))
  Z_rand = np.random.binomial(1, 0.8, size = 10)
  c.append(sum(Z_rand)/len(Z_rand))
  U_rand = np.random.binomial(1, 0.8, size = 30)
  d.append(sum(U_rand)/len(U_rand))
  V_rand = np.random.binomial(1, 0.8, size = 50)
  e.append(sum(V_rand)/len(V_rand))
  W_rand = np.random.binomial(1, 0.8, size = 100)
  f.append(sum(W_rand)/len(W_rand))

plt.hist(a, bins = 20, label = 'simulated', alpha = 0.5, density = True)
plt.hist(b, bins = 20, label = 'simulated', alpha = 0.5, density = True)
plt.hist(c, bins = 20, label = 'simulated', alpha = 0.5, density = True)
plt.hist(d, bins = 20, label = 'simulated', alpha = 0.5, density = True)
plt.hist(e, bins = 20, label = 'simulated', alpha = 0.5, density = True)
plt.hist(f, bins = 20, label = 'simulated', alpha = 0.5, density = True)
plt.show
```
